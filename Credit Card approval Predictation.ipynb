```python
import pandas as pd

cc_data=pd.read_csv("E:\ml data set\crx.data", header=None)

cc_data.head()
```

```python
cc_data_dis=cc_data.describe()
print(cc_data_dis)

print("\n")

cc_data_info=cc_data.info()
print(cc_data_info)

print("\n")

cc_data.tail(25)
```

```python
import numpy as np


print(cc_data.isnull().values.sum())

cc_data=cc_data.replace('?',np.nan)

print(cc_data.isnull().sum())

cc_data.tail(25)
```

```python
cc_data.fillna(cc_data.mean(),inplace=True)

print(cc_data.isnull().sum())
```

```python
for i in cc_data:
    if cc_data[i].dtype=='object':
        cc_data.fillna(cc_data[i].value_counts().index[0],inplace=True)
        
print('Total missing values:' + str(cc_data.isnull().values.sum()))
print('Missing values in each column:')
cc_data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for i in cc_data:
    if cc_data[i].dtypes=='object':
        le.fit(cc_data[i])
        cc_data[i]=le.transform(cc_data[i])
        
cc_data.info()
```

```python
from sklearn.model_selection import train_test_split

cc_data=cc_data.drop([11,13],axis=1)

print(cc_data.head())

cc_data=cc_data.values

X,y = cc_data[:,0:12] , cc_data[:,13]

X_train, X_test, y_train, y_test = train_test_split(X,
                                y,
                                test_size=0.33,
                                random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))

rescaledX_train = scaler.fit_transform(X_train)

rescaledX_test = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()

logreg.fit(rescaledX_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix

y_pred = logreg.predict(rescaledX_test)

print("Accuracy of logistic regression classifier: ", logreg.score(rescaledX_test, y_test))

print('Confusion matrix: \n ', confusion_matrix(y_test, y_pred))
```
